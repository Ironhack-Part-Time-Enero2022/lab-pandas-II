```python
#seleccionamos las columnas con las que vamos a trabajar y las renombramos para que sea más sencillo leerlas.

df_xlsx
df_resumen = df_xlsx[["espacio_prot_categoria", "espacio_prot_figura","espacio_prot_nombre","espacio_prot_superficie_ha","espacio_prot_fecha_declaracion"]]

df_resumen.rename(columns = {"espacio_prot_categoria":"categoria","espacio_prot_figura":"figura","espacio_prot_nombre":"nombre","espacio_prot_superficie_ha":"superficie","espacio_prot_fecha_declaracion":"fecha"}, inplace = True)
df_resumen
```

```python
#suma total de la superficie protegida en Madrid

df_resumen['superficie'].sum()
```

```python
#podemos comprobar que las últimas y más grandes declaración de espacios
#protegidos se han realizado desde 1988, compensandose aún alterando el Gobierno de la CCAA

# ordenar de may-menr

datos = df_resumen[["superficie", "fecha"]]

datos.sort_values(by=["superficie"], ascending = False)
```

# Crea alguna función y aplícala a tu dataframe usando `apply`

```python
#La media de la columna superficie por año para ver
#la media del crecimietno de la superficie protegida.

resultado = datos.groupby(["fecha"])["superficie"].mean()

resultado = resultado.reset_index()
```

```python
def limpiar_fecha(x):
    return x.year
```

```python
resultado["fecha_nueva"] = resultado["fecha"].apply(limpiar_fecha)
```
